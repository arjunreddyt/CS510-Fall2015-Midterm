```python
from attractor import Attractor
%matplotlib inline
```

```python
s=Attractor()
s.evolve()
s.ploty()
s.plotz()
s.plotxy()
s.plotyz()
s.plotzx()
s.plot3d()
```

Q)How does the plotted solution depend on your choice of time step size, and your choice of increment? Do you see an improvement in precision from using the higher-order integration methods?

Decreasing the time step size will increase the and accuracy of the Runge-Kutta model. A higher Runge-Kutta order will produce a more accurate results.
